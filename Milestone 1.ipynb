```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import requests
```

```python
positive_url = "https://ptrckprry.com/course/ssd/data/positive-words.txt"
negative_url = "https://ptrckprry.com/course/ssd/data/negative-words.txt"

payload={}
headers = {}

negative_response = requests.request("GET", negative_url, headers=headers, data=payload)
positive_response = requests.request("GET", positive_url, headers=headers, data=payload)

negative_list = negative_response.text.split('\n')[35:-1]
positive_list = positive_response.text.split('\n')[35:-1]
```

```python
positive_response = ['Good!', 'Great!', 'Excellent!', 'Awesome!', 'Fantastic!', 'Amazing!']
negative_response = ["Too Bad!","Womp,womp!",'bad', 'not good', 'terrible', 'awful', 'disappointing', 'displeased']
neutral_response = ["Hmm..","I see."]    
```

```python
data_store = pd.DataFrame()
```

```python
## Simple interactive chatbot
user_responses = []
user_name = input("What is your name? ")
print("Type 'end' or 'quit' to quit.")

while True:
    question = f"How's your day?, {user_name}"
    user_input = input(question)
    
    new_response = {'user_name': user_name, 'question': question, 'response': user_input}
    user_responses.append(new_response)    
    
    count_positive = 0
    count_negative = 0
    
    if ("end" in user_input) | ("quit" in user_input):
        break
    
    for i in user_input.split():
        if i in positive_list:
            count_positive += 1
        elif i in negative_list:
            count_negative += 1
#Case one        
    if count_positive > count_negative:
        question = f" What made your day good?"
        user_input = input(random.choice(positive_response)+ question + f"{user_name}")
        
        new_response = {'user_name': user_name, 'question': question, 'response': user_input}
        user_responses.append(new_response) 
        
        if ("end" in user_input) | ("quit" in user_input):
            break
        
        for i in user_input.split():
            if i in positive_list:
                count_positive += 1
            elif i in negative_list:
                count_negative += 1

        if count_positive > count_negative:
            question = f" Do you have something expectant tomorrow?"
            user_input = input(random.choice(positive_response)+ question + f"{user_name}")
            
            new_response = {'user_name': user_name, 'question': question, 'response': user_input}
            user_responses.append(new_response) 
            
            for i in user_input.split():
                if i in positive_list:
                    count_positive += 1
                elif i in negative_list:
                    count_negative += 1

            if count_positive > count_negative:
                print(random.choice(positive_response)+"It was great talking to you. Farewell!")    

            elif count_positive < count_negative:
                print(random.choice(negative_response)+"It was great talking to you. Farewell!")

            else:
                print(random.choice(neutral_response)+"It was great talking to you. Farewell!")

        elif count_positive < count_negative:
            question =  f" Do you have any ideas for improving your mood?"
            user_input = input(random.choice(negative_response)+ question + f"{user_name}")  
            
            new_response = {'user_name': user_name, 'question': question, 'response': user_input}
            user_responses.append(new_response) 
            
            for i in user_input.split():
                if i in positive_list:
                    count_positive += 1
                elif i in negative_list:
                    count_negative += 1

            if count_positive > count_negative:
                print(random.choice(positive_response)+"It was great talking to you. Farewell!")    

            elif count_positive < count_negative:
                print(random.choice(negative_response)+"It was great talking to you. Farewell!")

            else:
                print(random.choice(neutral_response)+"It was great talking to you. Farewell!")

        else:
            question = f" Hope you have a good day, do you have any other plan next? "
            user_input = input(random.choice(neutral_response)+ question + f"{user_name}")
            
            new_response = {'user_name': user_name, 'question': question, 'response': user_input}
            user_responses.append(new_response) 
    
            for i in user_input.split():
                if i in positive_list:
                    count_positive += 1
                elif i in negative_list:
                    count_negative += 1

            if count_positive > count_negative:
                print(random.choice(positive_response)+"It was great talking to you. Farewell!")    

            elif count_positive < count_negative:
                print(random.choice(negative_response)+"It was great talking to you. Farewell!")

            else:
                print(random.choice(neutral_response)+"It was great talking to you. Farewell!")

#Case two            
    elif count_positive < count_negative:
        question = f" What specifically happened?"
        user_input = input(random.choice(negative_response)+ question + f" {user_name}") 
        
        new_response = {'user_name': user_name, 'question': question, 'response': user_input}
        user_responses.append(new_response) 
    
        
        if ("end" in user_input) | ("quit" in user_input):
            break
        
        for i in user_input.split():
            if i in positive_list:
                count_positive += 1
            elif i in negative_list:
                count_negative += 1

        if count_positive > count_negative:
            question = f" That's pretty good, do you want to look forward to tomorrow's plans?"
            user_input = input(random.choice(positive_response)+ question + f" {user_name}")    
            
            new_response = {'user_name': user_name, 'question': question, 'response': user_input}
            user_responses.append(new_response) 
    
            for i in user_input.split():
                if i in positive_list:
                    count_positive += 1
                elif i in negative_list:
                    count_negative += 1

            if count_positive > count_negative:
                print(random.choice(positive_response)+"It was great talking to you. Farewell!")    

            elif count_positive < count_negative:
                print(random.choice(negative_response)+"It was great talking to you. Farewell!")

            else:
                print(random.choice(neutral_response)+"It was great talking to you. Farewell!")

        elif count_positive < count_negative:
            question = f" Do you have any ideas for improving your mood?"
            user_input = input(random.choice(negative_response)+ question + f"{user_name}")  
            
            new_response = {'user_name': user_name, 'question': question, 'response': user_input}
            user_responses.append(new_response) 
    
            for i in user_input.split():
                if i in positive_list:
                    count_positive += 1
                elif i in negative_list:
                    count_negative += 1

            if count_positive > count_negative:
                print(random.choice(positive_response)+"It was great talking to you. Farewell!")    

            elif count_positive < count_negative:
                print(random.choice(negative_response)+"It was great talking to you. Farewell!")

            else:
                print(random.choice(neutral_response)+"It was great talking to you. Farewell!")

        else:
            question = f" Hope you have a good day, do you have any other plan next? "
            user_input = input(random.choice(neutral_response)+ question + f"{user_name}")
            
            new_response = {'user_name': user_name, 'question': question, 'response': user_input}
            user_responses.append(new_response) 
    
            for i in user_input.split():
                if i in positive_list:
                    count_positive += 1
                elif i in negative_list:
                    count_negative += 1

            if count_positive > count_negative:
                print(random.choice(positive_response)+"It was great talking to you. Farewell!")    

            elif count_positive < count_negative:
                print(random.choice(negative_response)+"It was great talking to you. Farewell!")

            else:
                print(random.choice(neutral_response)+"It was great talking to you. Farewell!")
    
#Case three    
    else:
        question = f" Do you have anything else you'd like to share?"
        user_input = input(random.choice(neutral_response)+ question + f"{user_name}")
        new_response = {'user_name': user_name, 'question': question, 'response': user_input}
        user_responses.append(new_response) 
    
        if ("end" in user_input) | ("quit" in user_input):
            break
        
        for i in user_input.split():
            if i in positive_list:
                count_positive += 1
            elif i in negative_list:
                count_negative += 1

        if count_positive > count_negative:
            question = f" That's pretty good, do you want to look forward to tomorrow's plans?"
            user_input = input(random.choice(positive_response)+ question + f" {user_name}")

            new_response = {'user_name': user_name, 'question': question, 'response': user_input}
            user_responses.append(new_response) 
    
            for i in user_input.split():
                if i in positive_list:
                    count_positive += 1
                elif i in negative_list:
                    count_negative += 1

            if count_positive > count_negative:
                print(random.choice(positive_response)+"It was great talking to you. Farewell!")    

            elif count_positive < count_negative:
                print(random.choice(negative_response)+"It was great talking to you. Farewell!")

            else:
                print(random.choice(neutral_response)+"It was great talking to you. Farewell!")

        elif count_positive < count_negative:
            question = f" Do you have any ideas for improving your mood?"
            user_input = input(random.choice(negative_response)+ question + f"{user_name}")  
            
            new_response = {'user_name': user_name, 'question': question, 'response': user_input}
            user_responses.append(new_response) 
    
            for i in user_input.split():
                if i in positive_list:
                    count_positive += 1
                elif i in negative_list:
                    count_negative += 1

            if count_positive > count_negative:
                print(random.choice(positive_response)+"It was great talking to you. Farewell!")    

            elif count_positive < count_negative:
                print(random.choice(negative_response)+"It was great talking to you. Farewell!")

            else:
                print(random.choice(neutral_response)+"It was great talking to you. Farewell!")

        else:
            question = f" Hope you have a good day, do you have any other plan next? "
            user_input = input(random.choice(neutral_response)+ question + f"{user_name}")
            
            new_response = {'user_name': user_name, 'question': question, 'response': user_input}
            user_responses.append(new_response) 
    
            for i in user_input.split():
                if i in positive_list:
                    count_positive += 1
                elif i in negative_list:
                    count_negative += 1

            if count_positive > count_negative:
                print(random.choice(positive_response)+"It was great talking to you. Farewell!")    

            elif count_positive < count_negative:
                print(random.choice(negative_response)+"It was great talking to you. Farewell!")

            else:
                print(random.choice(neutral_response)+"It was great talking to you. Farewell!")
```
